```python
import tensorflow as tf
from tensorflow.keras import Model,Sequential,initializers
from tensorflow.keras.layers import Dense,Dropout
```

```python
tf.random.set_seed(42)## taking the seed as 42 
```

```python
initializer=tf.keras.initializers.GlorotUniform()
```

```python
audio_input=tf.keras.Input(shape=(audio_feat),name="audio_input")
model1=Dense(512,activation="relu",kernel_initializer=initializer)(audio_input)
model1=Dropout(0.2)(model1)

model1=Dense(256,activation="relu",kernel_initializer=initializer)(audio_input)
model1=Dropout(0.2)(model1)

model1=Dense(128,activation="relu",kernel_initializer=initializer)(audio_input)
model1=Dropout(0.2)(model1)

model1=Dense(64,activation="relu",kernel_initializer=initializer)(audio_input)
model1=Dropout(0.2)(model1)

model1=Dense(32,activation="relu",kernel_initializer=initializer)(audio_input)
model1=Dropout(0.2)(model1)

output_model_1=Dense(3,activation="softmax",kernel_initializer=initializer)(model1)
```

```python
final_model_1=Model(inputs=audio_input,outputs=output_model_1)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
ES=EarlyStopping(patience=10)
from tensorflow.keras.optimizers import Adam
optimizer=Adam()
final_model_1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer=optimizer,metrics=['accuracy'])
```

```python
history=final_model_1.fit(
    x=train_audio_features,   ##X_train is train_audio_features
    y=train_y_extra, ## only taking extraverasion
    validation_data=(val_audio_features,val_y_extra),## val x and val y (extraverasion)
    epochs=50,   
    callbacks=ES
)
```

```python
_,test_acc=final_model_1.evaluate(test_audio_features,test_y_extra)#xtest ,ytest
_,valid_acc=final_model_1.evaluate(val_audio_features,val_y_extra)#x validation y validation
```

```python
print("test acc is",test_acc*100)
print("validation acc  is",valid_acc*100)
```
